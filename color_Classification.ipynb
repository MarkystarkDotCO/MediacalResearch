
#  ML dataset setting.


```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
from sklearn import svm, metrics, datasets
from sklearn.utils import Bunch
from sklearn.model_selection import GridSearchCV, train_test_split

from skimage.io import imread
from skimage.transform import resize


def load_image_files(container_path, dimension):
    image_dir = Path(container_path)
    folders = [directory for directory in image_dir.iterdir() if directory.is_dir()]
    categories = [fo.name for fo in folders]

    descr = "A image classification dataset"
    images = []
    flat_data = []
    target = []
    for i, direc in enumerate(folders):
        for file in direc.iterdir():
            img = imread(file)
            img_resized = resize(img, dimension, anti_aliasing=True, mode='reflect')
            flat_data.append(img_resized.flatten())
            images.append(img_resized)
            target.append(i)
    flat_data = np.array(flat_data)
    target = np.array(target)
    images = np.array(images)

    return Bunch(data=flat_data,
                 target=target,
                 target_names=categories,
                 images=images,
                 DESCR=descr)
```

```python
image_dataset_train_aug1_128x128 = load_image_files("/content/augment1/train", dimension=(224, 224, 3))
image_dataset_test_aug1_128x128 = load_image_files("/content/augment1/test", dimension=(224, 224, 3))
```

```python
"V1 128x128"
X_train_aug1_128x128, _0, y_train_aug1_128x128, _0 = train_test_split(
    image_dataset_train_aug1_128x128.data, image_dataset_train_aug1_128x128.target, train_size=629, random_state=109)

X_test_aug1_128x128, _0, y_test_aug1_128x128, _0 = train_test_split(
    image_dataset_test_aug1_128x128.data, image_dataset_test_aug1_128x128.target, train_size=57,random_state=109)
```
